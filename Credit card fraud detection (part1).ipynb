```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('creditcard.csv')
df.head()
```

```python
df.head()
df.shape
df.info()
df.isnull().sum()
df.describe()
```

```python
df.columns
```

```python
df['class'].value_counts()
```

```python
df['Class'].value_counts()
```

```python
df['Class'].value_counts().plot(kind='bar',color=['skyblue','salmon'])
plt.title('Transaction Class Distribution')
plt.xticks([0,1],['Non-Fraud','Fraud'],rotation=0)
plt.ylabel('count')
plt.show()
```

```python
df['Amount'].describe()
```

```python
sns.boxplot(x='Class',y='Amount',data=df)
plt.title('Amount Distribution by Class')
plt.xticks([0,1],['Non-Fraud','Fraud'])
plt.show()
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(),cmap='coolwarm',annot=False)
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=df['Amount'])
plt.title('Amount Distribution')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=df['Amount'])
plt.title('Amount Distribution')
plt.show()          
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['credit_limit'],y=df['Amount'])
plt.title('Credit limit vs Amount')
plt.show()
```

```python
sns.scatterplot(x=df['Time'], y=df['Amount'], hue=df['Class'])
plt.title('Amount over Time (Fraud vs Non-Fraud)')
plt.xlabel('Time')
plt.ylabel('Amount')
plt.legend(title='Class')
plt.show()
```
